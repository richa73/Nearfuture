```python
import spacy
```

```python
nlp=spacy.blank("en")
```

```python
type(nlp)
```

```python
doc1=nlp('Today is Tuesday, 3rd Jan. We are startig the NLP course')
```

```python
type(doc1)
```

## Stop words

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
print(STOP_WORDS)
```

```python
len(STOP_WORDS)
```

```python
for token in doc1:
    print(token,'==>',token.is_stop)
```

```python
# Find the no of non-stop words tokens in the doc1

count=0
for token in doc1:
    if token.is_stop==False:
        print(token)
        count=count+1
print('The count of non -stop words tokens:',count)
    
```
