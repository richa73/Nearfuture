```python
def compare(dox):
    texts = [] #Collection of tokens

    for d in dox:
        text = []
        for t in d:
            if not t.is_stop and not t.is_punct and not t.like_num:
                text.append(t.lemma_)
        texts.append(text)

    #Creation of Corpus
    dict_1 = Dictionary(texts)

    #Bag of Words
    bow = []
    for i in texts:
        bow.append(dict_1.doc2bow(i))
    
    #Bag of Words in matrix format
    bow_matrix = matutils.corpus2dense(bow,num_terms = len(dict_1))

    #TFIDF
    tfidf = TfidfModel(bow)

    tfidf_vec = []
    for vec in bow:
        tfidf_vec.append(tfidf[vec])
    
    #Similarity of Document
    sim = MatrixSimilarity(tfidf_vec, num_features = len(dict_1))
    print(sim[tfidf_vec[0]])
    print(sim[tfidf_vec[1]])
```

```python
import nltk
import string
from heapq import nlargest
#nltk.download('stopwords')

def summary(text):
    if text.count(". ") > 20:
        length = int(round(text.count(". ")/10, 0))
    else:
        length = 1

    nopuch =[char for char in text if char not in string.punctuation]
    nopuch = "".join(nopuch)

    processed_text = [word for word in nopuch.split() if word.lower() not in nltk.corpus.stopwords.words('english')]

    word_freq = {}
    for word in processed_text:
        if word not in word_freq:
            word_freq[word] = 1
        else:
            word_freq[word] = word_freq[word] + 1

    max_freq = max(word_freq.values())
    for word in word_freq.keys():
        word_freq[word] = (word_freq[word]/max_freq)

    sent_list = nltk.sent_tokenize(text)
    sent_score = {}
    for sent in sent_list:
        for word in nltk.word_tokenize(sent.lower()):
            if word in word_freq.keys():
                if sent not in sent_score.keys():
                    sent_score[sent] = word_freq[word]
                else:
                    sent_score[sent] = sent_score[sent] + word_freq[word]

    summary_sents = nlargest(length, sent_score, key=sent_score.get)
    summary = " ".join(summary_sents)
    print(summary)
```
